# Train

```python
from sklearn.svm import SVC
```

```python
# Ø¢Ù…ÙˆØ²Ø´ SVC Ø®Ø·ÛŒ
svc_linear = SVC(kernel='linear', random_state=33, probability=True)
svc_linear.fit(X_train, y_train)
```

```python
# 1/||w|| Ù…Ø­Ø§Ø³Ø¨Ù‡
w_norm = np.linalg.norm(svc_linear.coef_)
print(f"1/||w|| = {w_norm:.4f}")
```

 ## ROC-AUC Ùˆ F1 ØŒRecall ØŒPrecision, Accuracy

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score, roc_curve ,precision_recall_curve, auc
```

```python
def evaluate_model(model, x, y, name='Dataset'):
  y_pred = model.predict(x)
  y_pred_prob = model.predict_proba(x)[:, 1]

  acc = accuracy_score(y, y_pred)
  f1 = f1_score(y, y_pred)
  recall = recall_score(y, y_pred)
  precision = precision_score(y, y_pred)
  roc_auc = roc_auc_score(y, y_pred_prob)

  print(f"\n--- {name} ---")
  print(f"Accuracy:  {acc:.4f}")
  print(f"F1 Score:  {f1:.4f}")
  print(f"Recall:    {recall:.4f}")
  print(f"Precision: {precision:.4f}")
  print(f"ROC-AUC:   {roc_auc:.4f}")

  return acc, f1, recall, precision, roc_auc
```

```python
acc_test, f1_test, recall_test, precision_test, roc_auc_test = evaluate_model(svc_linear, X_test, y_test, 'Test')
acc_val, f1_val, recall_val, precision_val, roc_auc_val = evaluate_model(svc_linear, X_val, y_val, 'Validation')
```

## Ù…Ø·Ø§Ù„Ø¹Ù‡ C

```python
C_values = [0.01, 0.1, 1, 10, 100, 1000]
results = []

for C in C_values:
    svc = SVC(kernel='linear', C=C, random_state=33, probability=True)
    svc.fit(X_train, y_train)

    # ØªØ¹Ø¯Ø§Ø¯ Support VectorÙ‡Ø§
    n_sv = svc.n_support_.sum()

    acc, f1, recall, precision, roc_auc = evaluate_model(svc, X_val, y_val, f'Test (C={C})')

    results.append({
        'C': C,
        'Support Vectors': n_sv,
        'Accuracy': acc,
        'F1': f1,
        'Recall': recall,
        'Precision': precision,
        'ROC-AUC': roc_auc
    })

# ØªØ¨Ø¯ÛŒÙ„ Ù†ØªØ§ÛŒØ¬ Ø¨Ù‡ DataFrame Ø¨Ø±Ø§ÛŒ Ù†Ù…Ø§ÛŒØ´
results_df = pd.DataFrame(results)
print(results_df)
```

## Ù†Ù…ÙˆØ¯Ø§Ø± Ù‡Ø§ÛŒ Ø§Ø±Ø²ÛŒØ§Ø¨ÛŒ

```python
best_idx = results_df['Accuracy'].idxmax()
best_C = results_df.loc[best_idx, 'C']
print(f"The best value of C is: {best_C}")
```

```python
best_svc = SVC(kernel='linear', C=best_C, random_state=33, probability=True)
best_svc.fit(X_train, y_train)
```

```python
y_test_pred = best_svc.predict(X_test)
y_test_prob = best_svc.predict_proba(X_test)[:, 1]
```

```python
# Ù…Ù†Ø­Ù†ÛŒ ROC
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)
roc_auc_val = auc(fpr, tpr)


plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_val:.3f})')
plt.plot([0,1], [0,1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve on Test Set')
plt.legend(loc="lower right")
plt.show()
```

```python
# Ù…Ù†Ø­Ù†ÛŒ Precision-Recall

precision, recall, pr_thresholds = precision_recall_curve(y_test, y_test_prob)
pr_auc_val = auc(recall, precision)

plt.figure(figsize=(8,6))
plt.plot(recall, precision, color='green', lw=2, label=f'PR curve (AUC = {pr_auc_val:.3f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve on Test Set')
plt.legend(loc='lower left')
plt.show()
```
