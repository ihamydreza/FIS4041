```python
features_to_map = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
data[features_to_map] = data[features_to_map].replace({'yes': 1, 'no': 0})
```

Ø±ÙˆØ´ IQR

```python
numeric_outliers = ['price', 'area']

for feature in numeric_outliers:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data = data[(data[feature] >= lower_bound) & (data[feature] <= upper_bound)]
```

## Function

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
```

```python

def evaluate_model_with_pca(model, X_train, X_test, y_train, y_test, components_list):
    """
    Evaluate a regression model with PCA dimensionality reduction.

    Parameters:
        model: regression model (e.g., Ridge(), LinearRegression())
        X_train, X_test: training and testing features
        y_train, y_test: target values
        components_list: list of PCA components to evaluate

    Returns:
        train_r2_scores, test_r2_scores: lists of R-squared scores
    """
    train_r2_scores = []
    test_r2_scores = []

    for n_components in components_list:
        # Apply PCA
        pca = PCA(n_components=n_components)
        X_train_pca = pca.fit_transform(X_train)
        X_test_pca = pca.transform(X_test)

        # Train the model
        model.fit(X_train_pca, y_train)

        # Predict
        y_train_pred = model.predict(X_train_pca)
        y_test_pred = model.predict(X_test_pca)

        # Calculate R-squared
        train_r2 = r2_score(y_train, y_train_pred)
        test_r2 = r2_score(y_test, y_test_pred)

        train_r2_scores.append(train_r2)
        test_r2_scores.append(test_r2)

        print(f"PCA with {n_components} components: Train R2 = {train_r2:.4f}, Test R2 = {test_r2:.4f}")

    return train_r2_scores, test_r2_scores
```

## Regression Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

degree = 2

pca = PCA(n_components=10)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

poly = PolynomialFeatures(degree=degree, include_bias=False)
X_train_poly = poly.fit_transform(X_train_pca)
X_test_poly = poly.transform(X_test_pca)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

y_train_pred = poly_model.predict(X_train_poly)
y_test_pred = poly_model.predict(X_test_poly)

train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
train_r2_poly = r2_score(y_train, y_train_pred)
test_r2_poly = r2_score(y_test, y_test_pred)

print("Train MSE:", train_mse)
print("Test MSE:", test_mse)
print("Train R2:", train_r2_poly)
print("Test R2:", test_r2_poly)
```

```python
import numpy as np

def extract_hidden_features_reg(model, X):

    h = X.copy()
    for i in range(len(model.coefs_) - 1):
        h = np.dot(h, model.coefs_[i]) + model.intercepts_[i]
        h = np.maximum(0, h)
    return h

X_train_features = extract_hidden_features_reg(mlp_model, X_train)
X_test_features = extract_hidden_features_reg(mlp_model, X_test)

print("Ø´Ú©Ù„ ÙˆÛŒÚ˜Ú¯ÛŒâ€ŒÙ‡Ø§ÛŒ Ø§Ø³ØªØ®Ø±Ø§Ø¬â€ŒØ´Ø¯Ù‡:", X_train_features.shape)
```
